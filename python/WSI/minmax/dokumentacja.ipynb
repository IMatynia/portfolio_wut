## Eksperyment 1 - porównanie głębokości
W tym eksperymencie porównane zostaną rozgrywki graczy o wybranych głębokościach przeszukiwania.

```python
from graphing_utils import graph_depth_comparisons
from constants import P_MIN, P_MAX
import os

jobs = {
    # (d_min, starting_player): [d_max_1, d_max_2, ..., d_max_m]
    (1, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (2, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (3, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (4, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (5, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (6, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (7, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (8, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
    (9, P_MIN): [1, 2, 3, 4, 5, 6, 7, 8, 9],
}
samples = 100
n = 3

graph_depth_comparisons(jobs, samples, n)
ret = os.system("notify-send 'experiment done'")

# Jeśli wykres zostanie utracony, jest on także zapisany w pliku ./3x3gridexperiments.png
```

### Wnioski
Z wykresów widać, że większa głębokość przeszukiwania daje większą szansę na wygraną z przeciwnikiem. W przypadku gdy obydwaj gracze mają głębokości większe niż 5, gra najczęściej się sprowadza do remisu ze względu na ich optymalną grę.

Należy też zauważyć, że to który gracz zaczyna ma duże znaczenie dla toku gry. Spójrzmy na przykład na D=8 VS D=3. Jeśli grę zaczyna gracz o D=8, wygrywać on będzie w około 80% grach, natomiast jeśli zacznie gracz D=3, prawdopodobieństwo to spadnie do poniżej 20%.
